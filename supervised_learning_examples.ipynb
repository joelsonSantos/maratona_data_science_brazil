```python
# Created by: Joelson AntÃ´nio dos Santos in December-20-2018
# This notebook was made for fun and can be improve for anyone that wish to use more sofisticated approaches.

# dataset (can use any other)
from sklearn.datasets import load_iris

# utils
import mglearn
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# data split approach
from sklearn.model_selection import train_test_split

# classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

############ training ###############################
def fitModel(algorithm, dataset):
    # train and test split approach (75% - 25%)
    X_train, X_test, y_train, y_test = train_test_split(dataset['data'], dataset['target'], random_state=0)
    if algorithm == "decision_tree":
        # fit model using decision tree
        model = DecisionTreeClassifier(random_state=0)
    elif algorithm == "naive_bayes":
        model = GaussianNB()
    elif algorithm == "random_forest":
        model = RandomForestClassifier(n_estimators=10)
    elif algorithm == "mlp":
        model = MLPClassifier(solver='lbfgs', alpha=0.7, hidden_layer_sizes=(5, 2), random_state=1)
    model.fit(X_train, y_train)
    y = model.predict(X_test)
    print(algorithm, ": ACC-> ", model.score(X_test, y_test))
```

```python
# loading dataset        
iris_dataset = load_iris()        
# creating a dataframe using pandas
iris_dataframe = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)
# ploting pair-attribute graphic
grr = pd.plotting.scatter_matrix(iris_dataframe, c=iris_dataset['target'], figsize=(15, 15), marker='o',
                       hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
## training and testing models        
fitModel(algorithm="decision_tree", dataset=iris_dataset)
fitModel(algorithm="naive_bayes", dataset=iris_dataset)
fitModel(algorithm="random_forest", dataset=iris_dataset)
fitModel(algorithm="mlp", dataset=iris_dataset)
```
